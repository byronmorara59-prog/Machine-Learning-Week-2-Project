```python
#Separating features and the target column
X_train = train.drop(columns=['id', 'yield'])
y_train = train['yield']

X_test = test.drop(columns=['id'])
test_ids = test['id']
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
for col in X_train.columns:
    plt.figure()
    plt.boxplot(X_train[col], vert=False)
    plt.title(f"Box Plot – {col}")
    plt.xlabel(col)
    plt.show();
```

```python
X_train_clean = X_train.copy()
y_train_clean = y_train.copy()

for col in X_train_clean.columns:
    Q1 = X_train_clean[col].quantile(0.25)
    Q3 = X_train_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    mask = (X_train_clean[col] >= lower) & (X_train_clean[col] <= upper)
    X_train_clean = X_train_clean[mask]
    y_train_clean = y_train_clean[mask]

X_train_clean = X_train_clean.reset_index(drop=True)
y_train_clean = y_train_clean.reset_index(drop=True)
```

```python
cat_cols = [col for col in X_train_clean.columns if X_train_clean[col].dtype == 'object']
print("Categorical columns:", cat_cols)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_clean)   
X_test_scaled  = scaler.transform(X_test)              
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

X_tr, X_val, y_tr, y_val = train_test_split(
    X_train_scaled, y_train_clean, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_tr, y_tr)

y_pred_lr = lr.predict(X_val)
mae_lr = mean_absolute_error(y_val, y_pred_lr)
print(f"Linear Regression MAE: {mae_lr:.4f}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

poly_model = Pipeline([
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('lr',   LinearRegression())
])

poly_model.fit(X_tr, y_tr)
y_pred_poly = poly_model.predict(X_val)
mae_poly = mean_absolute_error(y_val, y_pred_poly)
print(f"Polynomial Regression (deg 2) MAE: {mae_poly:.4f}")
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(X_tr, y_tr)

y_pred_ridge = ridge.predict(X_val)
mae_ridge = mean_absolute_error(y_val, y_pred_ridge)
print(f"Ridge Regression MAE: {mae_ridge:.4f}")
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=1.0)
lasso.fit(X_tr, y_tr)

y_pred_lasso = lasso.predict(X_val)
mae_lasso = mean_absolute_error(y_val, y_pred_lasso)
print(f"Lasso Regression MAE: {mae_lasso:.4f}")
```

```python
from sklearn.linear_model import ElasticNet

enet = ElasticNet(alpha=1.0, l1_ratio=0.5)
enet.fit(X_tr, y_tr)

y_pred_enet = enet.predict(X_val)
mae_enet = mean_absolute_error(y_val, y_pred_enet)
print(f"Elastic Net MAE: {mae_enet:.4f}")
```

```python
best_model = Ridge(alpha=1.0)
best_model.fit(X_train_scaled, y_train_clean)

test_predictions = best_model.predict(X_test_scaled)

submission = pd.DataFrame({
    'id':    test_ids.values,
    'yield': test_predictions
})

submission.to_csv('submission.csv', index=False)
submission.head(10)
```
